### 1.1.2 Colorspace Conversion

Now, you will learn how to convert images from one colorspace to another, like BGR ↔ Gray, BGR ↔ YUV etc. There are more than 150 color-space conversion methods available in OpenCV. 

**QUESTION:** (/4)

In the next two cells, implements the conversion function going from BGR image to GRAY image and BGR to YUV image. There are built-in functions in OpenCV doing exactly that but for the moment we'll ask you to implement your **own** function in order to get the idea of what is happening under the hood.

```python
def convert_bgr_to_grayscale(frame):
    """
    Convert a given bgr image into grayscale image
    :param frame: Color image to convert
    :return: Grayscale image as numpy array
    """
    
    # YOUR CODE HERE   

    b = frame[..., 0].astype(np.float32)
    g = frame[..., 1].astype(np.float32)
    r = frame[..., 2].astype(np.float32)

    gray = 0.114 * b + 0.587 * g + 0.299 * r  
    image = np.clip(gray, 0, 255)
    return image.astype(np.float32)
```

```python
def convert_bgr_to_yuv(frame):
    """
    Convert a given rgb image into hsv image
    :param frame: Color image to convert
    :return: YUV image as numpy array
    """
    
    # YOUR CODE HERE

    b = frame[..., 0].astype(np.float32)
    g = frame[..., 1].astype(np.float32)
    r = frame[..., 2].astype(np.float32)

    y = 0.114*b + 0.587*g + 0.299*r
    u = -0.14713*r - 0.28886*g + 0.436*b + 128.0
    v = 0.615*r - 0.51499*g - 0.10001*b + 128.0

    yuv = np.stack([y, u, v], axis=-1)
    image =  np.clip(yuv, 0, 255).astype(np.uint8)
    return image.astype(np.uint8)
```

```python
def threshold_bgr_image(frame, lower, upper):
    """
    Segment image.
        1. Convert image to HSV color space
        2. Find pixels in the desired color range (lower, upper)
        3. Apply the mask to the image. 
    :param frame: BGR image to segment
    :param lower: Lower threshold value
    :param upper: Upper threshold value
    :return res: Segmented image
    :return mask: binary mask
    """
    img_hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    img_mask = cv.inRange(img_hsv, lower, upper)
    mask = img_mask
    res = cv.bitwise_and(frame, frame, mask = mask)


        
    return res, mask
```

**Pros / cons (color vs grayscale):**

- **Color (HSV) segmentation**
  - **Pros:** isolates specific colors much better (threshold on Hue/Saturation/Value), typically gives a cleaner mask for yellow.
  - **Cons:** can be sensitive to illumination changes (shadows/highlights affect S and V), thresholds may need tuning.

- **Grayscale segmentation**
  - **Pros:** simpler and faster; works well when the object is distinguishable mainly by brightness (dark vs bright).
  - **Cons:** loses color information → cannot reliably separate objects with similar brightness; more false detections in scenes with multiple bright colors (as in this example).

### 1.1.3 Image Histogram

An histogram is a graph or plot, which gives you an overall idea about the intensity distribution of the pixels of the image. It is just another way of understanding the image. By looking at the histogram of an image, you get intuition about contrast, brightness, intensity distribution etc of that image. Almost all image processing tools today, provides features on histogram.

```python
def compute_image_histogram(image):
    """
    Compute image color distribution for each channels of a given `image`
    :param image: Image to extract histograms for
    :return: List of histogram, one per channel.
    """
    hist = []
    if image.ndim == 2:
        h = cv.calcHist([image.astype(np.uint8)], [0], None, [256], [0, 256])
        hist.append(h)
        return hist
    for ch in range(image.shape[2]):
        h = cv.calcHist([image.astype(np.uint8)], [ch], None, [256], [0, 256])
        hist.append(h)
        
   
    
    hist = np.array(hist)
    # YOUR CODE HERE


    return hist
```

**YOUR ANSWER** 

**Pixel intensity** represents the brightness value of a pixel.

For an 8-bit image:
- **0** → black (no light)
- **255** → white (maximum brightness)
- Values in between → shades of gray

For a color image (RGB/BGR):
- Each pixel has **three intensity values** (one per channel)
- These values determine how much red, green, and blue contribute to the final color

Therefore, pixel intensity quantifies brightness (or channel strength), and the histogram shows how these intensities are distributed across the image.

#### Histogram equalization
Consider an image whose pixel values are confined to some specific range of values only. For instance, brighter image will have all pixels confined to high values. But a good image will have pixels from all regions of the image. So you need to stretch this histogram to either ends and that is what Histogram Equalization does. This normally improves the contrast of the image ([doc](https://en.wikipedia.org/wiki/Histogram_equalization)).

**QUESTION:** (/3)

In the next section implement such contrast enhancement function (i.e. based on histogram equalization)

```python
def contrast_enhancement(image):
    """
    Compute histogram and apply histogram equalization on a given image.
    :param image: Image to enhance contrast
    :return: tuple: enhanced image, input image histogram, transformed image histogram
    """
    
    hist = compute_image_histogram(image)
    if image.ndim == 2:
        image_equ = cv.equalizeHist(image.astype(np.uint8))
    else:
        chans = []
        for ch in range(image.shape[2]):
            ch_eq = cv.equalizeHist(image[..., ch].astype(np.uint8))
            chans.append(ch_eq)
        image_equ = np.stack(chans, axis=2)

    hist_equ = compute_image_histogram(image_equ)
    
    return image_equ, hist, hist_equ
```

```python
# Load image
img = cv.imread(os.path.join('.','data', 'matterhorn_snow.jpg'), cv.IMREAD_GRAYSCALE)

# Enhance contrast with histogram equalization
img_equ, hist, hist_equ = contrast_enhancement(img)

# Display
fig, ax = plt.subplots(2, 2, figsize=(17, 9))
display_image(img, axes=ax[0][0])
ax[0][0].set_title('Original')
# Histogram original
ax[0][1].bar(np.arange(hist.shape[0]), hist.reshape(-1),  width=1)
ax[0][1].set_xlim([0, 256])
ax[0][1].set_title('Histogram')
# Histogram equalized
display_image(img_equ, axes=ax[1][0])
ax[1][0].set_title('Histogram Equalized')
ax[1][1].bar(np.arange(hist_equ.shape[0]), hist_equ.reshape(-1),  width=1)
ax[1][1].set_xlim([0, 256])
ax[1][1].set_title('Histogram')
plt.show()
```

- The contrast increased.

#### 2. How did the histogram change?

**QUESTION:** (/2)

direction of filtering (`x` or `y`), and the filter size  and returns the filtered image. Use the corresponding OpenCV function with the parameter `ddepth=cv.CV_64F`. 


```python
def apply_sobel_filter(img, direction, filter_size):
    
    """
    Apply the sobel filter on a given image.
    :param img: Image to enhance contrast
    :param direction: Direction of the derivative
    :param filter_size: Size of the filter
    :return: img_filt: filtered image
    """
    
    dx, dy = (1, 0) if direction == "x" else (0, 1)

    img_filt = cv.Sobel(img, ddepth=cv.CV_64F, dx=dx, dy=dy, ksize=filter_size)

    return img_filt
```

#### Otsu Binarization

In global thresholding, we used an arbitrary value for threshold value. How can we know a value we selected is good or not? Answer is, trial and error method. But consider an image whose histogram has two peaks (a bimodal image). For that image, we can approximately take a value in the middle of those peaks as threshold value. That is what Otsu binarization does. In simple words, it automatically calculates a threshold value from image histogram for a bimodal image. (For images which are not bimodal, binarization won’t be accurate.)

For this, our `cv.threshold()` function is used, but pass an extra flag, `cv.THRES_OTSU`. For threshold value, simply pass zero. Then the algorithm finds the optimal threshold value and returns you as the second output, retVal. If Otsu thresholding is not used, retVal is same as the threshold value you used.

### 1.3.1 Introduction

Histogram of Oriented Gradients (*HOGs*) are descriptors mainly used in computer vision and machine learning for object detection. HOG features were first introduced by Dalal and Triggs in their CVPR 2005 paper: "Histogram of Oriented Gradients for Human Detection" (<http://ieeexplore.ieee.org/document/1467360/>); which transforms image pixels into a vector representation that is sensitive to broadly informative image features regardless of confounding factors like illumination. Later on, we will use these descriptors for detection and tracking. 


In their work, Dalal and Triggs proposed HOG and a 5-stages descriptor to classify humans in static images.
The 5 stages included:

    1. Pre-processing and scaling.
    2. Computing gradients in both the x and y directions.
    3. Obtaining weighted votes in spatial cells (local histograms).
    4. Contrast normalizing overlapping spatial cells (Blocks).
    5. Collecting all Histograms of Oriented gradients to form the final feature vector.


In most real-world applications, HOG is used in conjunction with a [__Linear SVM__](https://en.wikipedia.org/wiki/Support_vector_machine) to perform object detection. HOG rapidly became one of the most used descriptors in image classification. The reason HOG is employed so heavily is because the local object appearance and shape can be characterized using the distribution of local intensity gradients. 

We’ll be discussing the steps necessary to combine both HOG and a Linear SVM into an object classifier later in this course. But for now, just understand that HOG is mainly used as a descriptor for object detection and that later these descriptors can be fed into a machine learning classifier.

### 1.3.4 Gradient computation

https://en.wikipedia.org/wiki/Image_gradient

Gradient vectors (or “image gradients”) are one of the most fundamental concepts in computer vision; many vision algorithms involve computing gradient vectors for each pixel in an image. A gradient vector it’s simply a measure of the change in pixel values along the x-direction and the y-direction around each pixel. The magnitude is defined as:

$$
G(x,y) = \sqrt{(\Delta x^2 + \Delta y^2)}
$$

and the phase: 

$$
\theta(x,y) = atan(\frac{\Delta x}{\Delta y})
$$

Where: $\Delta x = f(x+1,y) - f(x-1,y)$ and $\Delta y = f(x,y+1) - f(x,y-1)$, are simply the directional change from one pixel to the other.

The simplest implementation of the aforementioned operators computes each value by using a mask operator: $-1|0|1$ over the pixel position in each direction. However the Sobel operator you implemented in Section 1.2.2is usually employed as it is more robust to intensity changes.

### Do the images make sense?

Yes, they do.  
The two images represent the first-order derivatives of the image:

- **Derivative along X direction** → highlights **vertical edges**.
- **Derivative along Y direction** → highlights **horizontal edges**.

This is because:
- The X derivative measures intensity changes **from left to right**.
- The Y derivative measures intensity changes **from top to bottom**.

Where intensity changes rapidly (edges), the derivative magnitude is high, so edges appear clearly.

---

### What do the brightest and darkest pixels mean?

The derivative images contain **positive and negative values**:

- **Bright pixels (positive values)**  
  → intensity increases in that direction.

- **Dark pixels (negative values)**  
  → intensity decreases in that direction.

- **Gray pixels (values near 0)**  
  → little or no intensity change (flat regions).

So:

- In the **X derivative image**, bright/dark edges correspond to vertical transitions (left-right changes).
- In the **Y derivative image**, bright/dark edges correspond to horizontal transitions (top-bottom changes).

In summary, the images show directional edge information, and the sign indicates whether the intensity is increasing or decreasing along that direction.

**QUESTION:** (/5)

In the next cell, implement the function `computeCellHistograms()` that takes as input the phase of the image, the number of bins, and the cell size, and returns a [list](https://docs.python.org/3/tutorial/datastructures.html) containing the histograms. 

```python
def computeCellHistograms(phase, cell_size, num_of_bins):
    """
    Compute histogram for each cells and concatenate them.
    
    :param phase: 1 channel image with the gradient direction.
    :params cell_size: cell size (8px) 
    :params num_of_bins: number of bins in the histogram
    :return concatenated_histograms: list of histograms
    """
    
    # Resulting histogram list
    concatenated_histograms =[];
    w = phase.shape[0]
    h = phase.shape[1]
    
    for i in range(w // cell_size):
        for j in range(h // cell_size):
            y0 = i * cell_size
            x0 = j * cell_size
            cell = phase[y0:y0+cell_size, x0:x0+cell_size].astype(np.float32)

            hist = cv.calcHist([cell], channels=[0], mask=None,
                               histSize=[num_of_bins], ranges=[0, 360])
            concatenated_histograms.append(hist)
     
    return concatenated_histograms
```

The function `computeCellHistograms()` computes the histogram of each cell. However, it assumes that each direction has the same importance and contributes equally to the histogram. In the original paper, they proposed to weight the histogram, for instance, using the gradient magnitude. The histogram bins of each cell are weighted by the gradient magnitude. **QUESTION:** (/5)

In the next cell, implement the function `computeCellWeightedHistogram()` to return weighted histograms. The new version takes as input the phase of the image, the cell size, the number of bins, and the gradient magnitude and returns a [list](https://docs.python.org/3/tutorial/datastructures.html) containing the histograms. 

**HINT:** Avoid pre-defined functions to compute the histogram and code your own function.



```python
import numpy as np

def computeCellWeightedHistogram(phase, magnitude, cell_size, num_of_bins):
    """
    Compute weighted histogram for each cell and concatenate them.

    :param phase: 2D array with gradient direction (degrees 0–360 or 0–180)
    :param magnitude: 2D array with gradient magnitude
    :param cell_size: size of each cell (e.g., 8)
    :param num_of_bins: number of bins
    :return: list of weighted histograms (1D numpy arrays)
    """

    weighted_and_concatenated_histograms = []

    H, W = phase.shape
    bin_width = 360 / num_of_bins   

    for i in range(H // cell_size):
        for j in range(W // cell_size):

            y0 = i * cell_size
            x0 = j * cell_size

            cell_phase = phase[y0:y0+cell_size, x0:x0+cell_size]
            cell_magnitude = magnitude[y0:y0+cell_size, x0:x0+cell_size]

            hist = np.zeros(num_of_bins)

            for y in range(cell_size):
                for x in range(cell_size):

                    angle = cell_phase[y, x]
                    mag = cell_magnitude[y, x]

                    bin_idx = int(angle // bin_width)

                    # safety for angle == 360
                    if bin_idx >= num_of_bins:
                        bin_idx = num_of_bins - 1

                    hist[bin_idx] += mag   # weight by magnitude

            weighted_and_concatenated_histograms.append(hist)

    return weighted_and_concatenated_histograms
```

### How can we see the effect of weighting from the two histograms?

The difference is visible by looking at the **height of the bins** and what they represent.

#### Left (without weighting)
- The y-axis represents the **number of pixels** in each orientation bin.
- Every pixel contributes equally, regardless of edge strength.
- Weak gradients and strong gradients have the same influence.
- The histogram reflects only how frequently orientations appear.

#### Right (with magnitude weighting)
- The y-axis represents the **sum of gradient magnitudes** in each bin.
- Strong edges contribute more than weak edges.
- Bins corresponding to dominant edge directions become significantly taller.
- Bins associated with weak/noisy gradients contribute less.

#### What we observe
- Some bins increase much more in the weighted version → those orientations correspond to **strong edges**.
- Less important orientations become relatively smaller.
- The weighted histogram emphasizes the **dominant structural directions** in the image.

In summary, weighting changes the histogram from a simple count of orientations to a measure of **how strong each orientation is**, making the representation more meaningful.

### 1.3.6 Block normalization

In the previous steps, we created the oriented histogram based on the gradient magnitude. However, gradients are sensitive to overall lighting. If you make the image darker by dividing all pixel values by 2, for example, the gradient magnitude will change by half, and, therefore, the histogram values will change by half. Ideally, we want our descriptor to be independent of lighting variations. In other words, we would like to “normalize” the histograms to stabilitze the results. 

Rather than normalizing each histogram individually, we group the cells into __blocks__ and normalize the histogram of the block instead of the indivual cell. These procedure makes the feature detector less sensitive to intensity changes.


<img src = "https://gurus.pyimagesearch.com/wp-content/uploads/2015/03/hog_contrast_normalization.gif">
<em> Image taken from: https://gurus.pyimagesearch.com </em>

Image above shows each block as a group of 4 cells (2 by 2 cells). The blocks overlap each other and, therefore, each cell is represented in the histograms multiple times. This redundancy actually improves accuracy. Finally, we take the resulting block histograms, concatenate them, and treat them as our final feature vector.


**QUESTION:** (/6)

In the next cell, implement the function `BlockNormalizationAndFeatureConcatenation()` that takes as input the histograms, and the image size and returns the feature vector of histogram normalized with the L2 norm. 

```python
def BlockNormalizationAndFeatureConcatenation(histograms, img_size, cell_size, NB_BLOCK_CELLS_X = 2, NB_BLOCK_CELLS_Y = 2):
    
    """
    Compute histogram  block normalization
    
    :param histograms: list of weighted histograms
    :params img_size:  size of the image
    :return 
    """
    
    feature_vector    = []

    H, W = img_size
    nb_cells_y = H // cell_size
    nb_cells_x = W // cell_size

    histograms = np.array(histograms)
    num_bins = histograms.shape[1]
    histograms = histograms.reshape(nb_cells_y, nb_cells_x, num_bins)

    feature_vector = []

    for i in range(nb_cells_y - NB_BLOCK_CELLS_Y + 1):
        for j in range(nb_cells_x - NB_BLOCK_CELLS_X + 1):

            block = histograms[i:i+NB_BLOCK_CELLS_Y,
                               j:j+NB_BLOCK_CELLS_X,
                               :]

            block_vector = block.flatten()

            norm = np.linalg.norm(block_vector, ord=2)
            block_vector = block_vector/(norm+1e6)
  

            feature_vector.append(block_vector)
    
    feature_vector = np.array(feature_vector)        
    return feature_vector
```

### Explanation of `visualize_hog_on_image(...)`

This function visualizes **HOG features** by drawing, for each **cell**, a short line that represents the **dominant gradient orientation** in that cell.

---

#### Inputs
- `img_gray`: grayscale image of shape $(H, W)$.
- `cell_histograms`: list of histograms, one per cell (each histogram has `num_bins` bins).  
  Usually these are **weighted histograms** (weighted by gradient magnitude).
- `cell_size`: size of each cell in pixels (e.g. `8`).
- `num_bins`: number of orientation bins (e.g. `9`).
- `angle_range`: range of angles:
  - `180` if orientations are wrapped to $[0,180)$ (typical HOG)
  - `360` if orientations are in $[0,360)$

---

#### Step 1 — Compute how many cells there are
The image is split into a grid of cells:

$$
N_y = \left\lfloor \frac{H}{\text{cell\_size}} \right\rfloor,\quad
N_x = \left\lfloor \frac{W}{\text{cell\_size}} \right\rfloor
$$

Then the list of histograms is reshaped into a 3D array:

$$
\text{hists} \in \mathbb{R}^{N_y \times N_x \times \text{num\_bins}}
$$

So each cell has its own histogram.

---

#### Step 2 — Convert bin indices into angles
Each histogram bin corresponds to an interval of angles.
The bin width is:

$$
\Delta\theta = \frac{\text{angle\_range}}{\text{num\_bins}}
$$

The function uses the **center of each bin**:

$$
\theta_k = \left(k + 0.5\right)\Delta\theta
$$

and converts degrees to radians (needed for `cos` and `sin`).

---

#### Step 3 — Find the dominant orientation per cell
For each cell histogram:
- `k = argmax(hist)` selects the bin with the largest value (dominant orientation).
- The line length is proportional to the dominant bin value:

$$
\text{strength} = \frac{\text{hist}[k]}{\max(\text{hists})}
$$

This normalizes values so they are between 0 and 1.

---

#### Step 4 — Draw a line in each cell
The line is drawn at the **center** of the cell:

$$
x_0 = c_x \cdot \text{cell\_size} + \frac{\text{cell\_size}}{2},\quad
y_0 = c_y \cdot \text{cell\_size} + \frac{\text{cell\_size}}{2}
$$

Its length is:

$$
L = \frac{\text{cell\_size}}{2} \cdot \text{strength}
$$

The direction is determined by:

$$
dx = L\cos(\theta_k),\quad dy = L\sin(\theta_k)
$$

Finally, the function plots a symmetric line segment:

- from $(x_0-dx,\; y_0-dy)$ to $(x_0+dx,\; y_0+dy)$

This gives a “glyph” showing the main edge direction in each cell.

---

#### Interpretation
- **Longer line** → stronger gradient / stronger edges in that cell.
- **Line direction** → dominant edge orientation in that cell.
- This provides an intuitive picture of what the HOG descriptor is capturing.

```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_hog_on_image(img_gray, cell_histograms, cell_size, num_bins, angle_range=180):
    """
    Visualize HOG-style features by drawing one dominant orientation per cell.

    img_gray: (H,W) grayscale image
    cell_histograms: list of histograms, one per cell (length=num_bins)
                     (use weightedHistograms for nicer visualization)
    cell_size: e.g. 8
    num_bins: e.g. 9
    angle_range: 180 if phase wrapped to [0,180), else 360
    """
    H, W = img_gray.shape
    nb_cells_y = H // cell_size
    nb_cells_x = W // cell_size

    hists = np.array(cell_histograms).reshape(nb_cells_y, nb_cells_x, num_bins)

    # bin center angles in radians
    bin_width = angle_range / num_bins
    angles = (np.arange(num_bins) + 0.5) * bin_width
    angles_rad = np.deg2rad(angles)


    max_val = hists.max()

    plt.figure(figsize=(8, 8))
    plt.imshow(img_gray, cmap="gray")
    plt.title("HOG-like visualization (dominant orientation per cell)")
    plt.axis("off")

    for cy in range(nb_cells_y):
        for cx in range(nb_cells_x):
            hist = hists[cy, cx]

            k = np.argmax(hist)           # dominant bin
            strength = hist[k] / max_val  # 0..1

            theta = angles_rad[k]

            # cell center
            y0 = cy * cell_size + cell_size / 2
            x0 = cx * cell_size + cell_size / 2

            # line length proportional to strength
            L = (cell_size / 2) * strength

            dx = L * np.cos(theta)
            dy = L * np.sin(theta)

            # draw line (both directions for symmetry)
            plt.plot([x0 - dx, x0 + dx], [y0 - dy, y0 + dy], linewidth=1)

    plt.show()
```
